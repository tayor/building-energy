# Estimation of Energy Performance of Residential Buildings Using Artificial Neural Networks

### Dataset description
+ X1 Relative Compactness
+ X2 Surface Area
+ X3 Wall Area
+ X4 Roof Area
+ X5 Overall Height
+ X6 Orientation
+ X7	Glazing Area 
+ X8	Glazing Area Distribution 
+ y1	Heating Load 
+ y2	Cooling Load

```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
def load_data():
    print('Loading data...')
    data=pd.read_csv('ENB2012_data.csv').as_matrix()
    X=data[:,0:8]
    y=data[:,8:10]
    return X,y
```

```python
def dnn():
    model = Sequential()
    model.add(Dense(8,input_dim=8,init='normal'))
    model.add(Dense(4))
    model.add(Dense(2))
    model.compile(loss='mean_squared_error',optimizer='rmsprop')
    return model
```

```python
seed = 7
np.random.seed(seed)
X, y= load_data()
# evaluate model with standardized dataset
estimator = KerasRegressor(build_fn=dnn, nb_epoch=10, batch_size=5, verbose=1)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(estimator, X, y, cv=kfold)
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```
